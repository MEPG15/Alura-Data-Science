### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación





```python
# Calculamos la facturación total sumando la columna 'Precio'
facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

# Mostramos los resultados
print(f"Facturación Tienda 1: {facturacion_tienda1}")
print(f"Facturación Tienda 2: {facturacion_tienda2}")
print(f"Facturación Tienda 3: {facturacion_tienda3}")
print(f"Facturación Tienda 4: {facturacion_tienda4}")

# Identificamos la que menos factura para darle el primer dato al señor Juan
tiendas_facturacion = {
    'Tienda 1': facturacion_tienda1,
    'Tienda 2': facturacion_tienda2,
    'Tienda 3': facturacion_tienda3,
    'Tienda 4': facturacion_tienda4
}

tienda_menor_ingreso = min(tiendas_facturacion, key=tiendas_facturacion.get)
print(f"\n--- Alerta ---")
print(f"La tienda con menor facturación es: {tienda_menor_ingreso}")
```

```python
import matplotlib.pyplot as plt

# Creamos la gráfica
plt.bar(tiendas_facturacion.keys(), tiendas_facturacion.values(), color=['blue', 'green', 'orange', 'red'])
plt.title('Comparativa de Facturación por Tienda')
plt.ylabel('Ingresos Totales')
plt.show()
```

# 2. Ventas por categoría

```python
# 2. Ventas por categoría (Corregido)
col_categoria = 'Categoría del Producto'

print("--- Top 5 Categorías más vendidas ---")
print(f"\nTienda 1:\n{tienda[col_categoria].value_counts().head(5)}")
print(f"\nTienda 2:\n{tienda2[col_categoria].value_counts().head(5)}")
print(f"\nTienda 3:\n{tienda3[col_categoria].value_counts().head(5)}")
print(f"\nTienda 4:\n{tienda4[col_categoria].value_counts().head(5)}")
```

# 3. Calificación promedio de la tienda


```python
# 3. Calificación promedio

# Calculamos el promedio de cada una
prom_t1 = tienda['Calificación'].mean()
prom_t2 = tienda2['Calificación'].mean()
prom_t3 = tienda3['Calificación'].mean()
prom_t4 = tienda4['Calificación'].mean()

print(f"Calificación Promedio (1 a 5 estrellas):")
print(f"Tienda 1: {prom_t1:.2f} ⭐")
print(f"Tienda 2: {prom_t2:.2f} ⭐")
print(f"Tienda 3: {prom_t3:.2f} ⭐")
print(f"Tienda 4: {prom_t4:.2f} ⭐")
```

# 4. Productos más y menos vendidos

```python
# 4. Productos más y menos vendidos
col_prod = 'Producto'

# Ejemplo para Tienda 4 (que es la que estamos vigilando)
print("--- Tienda 4: Productos estrella ---")
print(tienda4[col_prod].value_counts().head(3))

print("\n--- Tienda 4: Productos menos vendidos ---")
print(tienda4[col_prod].value_counts().tail(3))
```

# 5. Envío promedio por tienda

```python
# 5. Envío promedio
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo promedio de envío:")
print(f"Tienda 1: ${envio_t1:.2f}")
print(f"Tienda 2: ${envio_t2:.2f}")
print(f"Tienda 3: ${envio_t3:.2f}")
print(f"Tienda 4: ${envio_t4:.2f}")
```
